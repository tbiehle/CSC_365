**Question 1:**
Report the current health status for all 20-year old students from Gabriel Pereira ('GP') high school who are in a romantic relationship (report the student Id and their health status).

```python
import pandas as pd
import numpy as np
df = pd.read_csv("Maths.csv")
df['Id'] = df.index

target_school = "GP"
target_age = 20
target_romantic = "yes"
target_health_col = "health" 

found_records = df[(df['school'] == target_school) &
                    (df['romantic'] == target_romantic) &
                    (df['age'] == target_age)]

if found_records.empty:
  print("No records were found that match the specified criteria.")
else:
  print("Health status for all 20-year old students from Gabriel Pereira high school who are in a romantic relationship.\n")

  health_values = ["very bad", "bad", "average", "good", "very good"]
  for index, row in found_records.iterrows():
    print(f"Student ID: {index}, Student Health: {row[target_health_col]} ({health_values[row[target_health_col] - 1]})")
```

**Question 3:** Find the highest total score a student from _Gabriel Pereira_ school received in their math class (add up _G1_, _G2_, and _G3_ scores). Report the score and the record ID.

```python
import pandas as pd
import numpy as np
df = pd.read_csv("Maths.csv")
df['Id'] = df.index

target_school = "GP"

GP = df[df["school"] == target_school]

max_score = float('-inf')
max_score_id = 0
for index, row in GP.iterrows():
  student_score = row["G1"] + row["G2"] + row["G3"]
  new_score = max(student_score, max_score)

  if new_score > max_score:
    max_score = new_score
    max_score_id = row['Id']

  
print("Highest total score a student from Gabriel Pereira school received in their math class:")
print(f"Student ID: {max_score_id}, Score: {max_score}")
```

**Question 4** Find all the students from  _Gabriel Pereira_ High School who had the top 10% highest score (sum of G1, G2 and G3) in their math class. Report their Ids, the G1, G2, and G3 scores, and their weekend and weekday alcohol consumption.


```python
import pandas as pd
import numpy as np
df = pd.read_csv("Maths.csv")
df['Id'] = df.index

target_school = "GP"

GP = df[df['school'] == target_school].copy()
GP['total_score'] = GP['G1'] + GP['G2'] + GP['G3']

GP = GP.sort_values('total_score', ascending=False) # GP now sorted by total score in descending order

top_ten_percent = GP.iloc[:int(0.1 * len(GP))] # gets first 10% of sorted data, resulting in top 10%


print("Student Id: G1, G2, G3, Weekend alcohol consumption, Weekday alcohol consumption")
for stud in top_ten_percent.itertuples():
  print(f"{stud.Id}: {stud.G1}, {stud.G2}, {stud.G3}, {stud.Walc}, {stud.Dalc}")
```

**Question 5:** Compare the average performance of female students who are engaged in extra curricular activities at _Gabriel Pereira_ and _Mousinho da Silveira_ schools in the math class (the G3 score). Report the full name (_Gabriel Pereira_ or _Mousinho da Silveira_) of the school with the higher average performance.

**Question 6:** Break the students into two categories by whether or not they are in a romantic relationship. For each group report (a) the number of students in the group, (b) the average math score (G3). Report the groups in descending order by the average score. (your output should contain two rows, and three values in each row: "yes" or "no" for relationship status, the number of students, and the average score).


```python
import pandas as pd
import numpy as np
df = pd.read_csv("Maths.csv")
df['Id'] = df.index

in_relationship = df[df['romantic'] == 'yes'] # group of students in relationship
no_relationship = df[df['romantic'] == 'no'] # group of students not in relationship

results = []
for group in [in_relationship, no_relationship]:
  romantic = group['romantic'].iloc[0]
  mean = group['G3'].mean()
  results.append([romantic, len(group), mean])

results.sort(key=lambda x: x[2], reverse=True)

print("In a relationship?")
for row in results:
    print(f"{row[0]}: Number of students in group: {row[1]}, average math score: {float(row[2]).__round__(2)}")


```

**Question 7:** Find the percent of male and female students in each school that consume considerable quantities of alcohol (value of 3 or above) _both_ on weekends and weekdays.

**Question 8:**  Find who reports better health on average - students with high math scores (16 or above), or students with low math scores.  Report the two averages.

```python
import pandas as pd
import numpy as np
df = pd.read_csv("Maths.csv")
df['Id'] = df.index

hlth_high = df[df['G3'] >= 16]['health'].mean()
hlth_low = df[df['G3'] < 16]['health'].mean()

print(f"Average health of students with math scores 16 or above: {hlth_high.__round__(2)}")
print(f"Average health of students with math scores lower than 16: {hlth_low.__round__(2)}")

if hlth_high > hlth_low:
  print("Students with high math scores have better health on average.")
else:
  print("Students with low math scores have better health on average.")
```

**Question 9:** Find whether the **percentage** of students with high consumption of alcohol (>=4 either on weekend or weekday) is higher among the students whose parents live together or separately (the "Pstatus" column).  Report the status with the higher percentage and the percentage itself.

```python
import pandas as pd
import numpy as np
df = pd.read_csv("Maths.csv")
df['Id'] = df.index

n_together = len(df[df['Pstatus'] == 'T'])
n_apart = len(df[df['Pstatus'] == 'A'])
stud_high_alc = df[(df['Dalc'] >= 4) | (df['Walc'] >= 4)]
p_together = stud_high_alc[stud_high_alc['Pstatus'] == 'T']
p_apart = stud_high_alc[stud_high_alc['Pstatus'] == 'A']

percent_t = (len(p_together) / n_together) * 100
percent_a = (len(p_apart) / n_apart) * 100

if percent_t > percent_a:
  print("The percentage of students with high consumption of alcohol is higher among those\n"
        "whose parents live together than those whose parents live apart.\n"
        f"Percentage: {percent_t.__round__(2)}%")
else:
  print("The percentage of students with high consumption of alcohol is higher among those\n"
        "whose parents live apart than those whose parents live together.\n"
        f"Percentage: {percent_a.__round__(2)}%")
```

**Question 10:** List the average alcohol consumption level (average between their weekend and weekday consumption) for each student with the highest G3 score in the class. Your output shall contain two values in each row: the Id of the student and their average alcohol consumption level.

```python
import pandas as pd
import numpy as np
df = pd.read_csv("Maths.csv")
df['Id'] = df.index

max_score = df['G3'].max()
students = df[df['G3'] == max_score]

print(f"Average alcohol consumption level for each student with highest G3 score in the class ({max_score}):")
for student in students.itertuples():
  print(f"Student ID: {student.Id}, Average alcohol consumption: {(student.Dalc + student.Walc) / 2}")
```
